**Importación y Carga de Librerias**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt; plt.style.use('ggplot')
%matplotlib inline
import warnings
# Ignoramos todos los 'warnings'
warnings.filterwarnings("ignore")
```

**Carga de Datos**

```python
# Carga de los datos del archivo 'homicidios_hechos' y visualizamos los datos
df_homicidios_hechos = pd.read_csv('homicidios_hechos.csv')
df_homicidios_hechos
```

```python
# Carga de los datos del archivo 'homicidios_victimas' y visualizamos los datos
df_homicidios_victimas = pd.read_csv('homicidios_victimas.csv')
df_homicidios_victimas
```

```python
# Carga de los datos del archivo 'lesiones' y visualización de los mismos
df_lesiones = pd.read_csv('lesiones_clean.csv')
df_lesiones
```

**Análisis y Graficación**

→ Para datos de 'homicidios'

```python
# Filtramos por registros 'sd' en estas columnas 'acusado' y 'victima'
data_sd = df_homicidios_hechos[(df_homicidios_hechos['acusado'] == 'sd') & (df_homicidios_hechos['victima'] == 'sd')]

# Visualizamos resultados de registros 'sd' (sin datos)
data_sd
```

```python
# Visualizamos cantidad de homicidios vinculado al número de victimas reportado
victims=list(df_homicidios_hechos["n_victimas"].unique())
cant=[]
for n in victims:
    cant.append(df_homicidios_hechos[df_homicidios_hechos["n_victimas"]==n]["n_victimas"].count())

pd.DataFrame({"n_victimas":victims,"cantidad":cant})
```

```python
# Agrupamos por año ('AAAA') mientras realizamos la sumatoria de victimas
victims_year = df_homicidios_hechos.groupby('aaaa')['n_victimas'].sum()

# Ordenamos cada año en base a cantidad y le damos una identidad (color)
sorted_years = victims_year.sort_values(ascending=True)
color_selected = sns.color_palette("Purples", n_colors=len(victims_year))
color_select = {year: color_selected[i] for i, year in enumerate(sorted_years.index)}

# Graficamos los datos
plt.figure(figsize=(16, 8))
bars = plt.bar(victims_year.index, victims_year, 
               color=[color_select[year] for year in victims_year.index])

# Caracteristicas del grafico
plt.xlabel('Año')
plt.ylabel('Total Victimas')
plt.title('Total de Homicidios por Año - Homicidios Hechos')
plt.xticks(rotation=0)
plt.grid(axis='y')

# Generaramos historial de colores
colors = [plt.Rectangle((0, 0), 1, 1, color=color_select[year]) for year in sorted_years.index]
plt.legend(colors, sorted_years.index, title='Año x N° de Victimas, Orden Descendiente')

plt.show()
```

```python
# Filtramos los datos para los años 2016 a 2021
filter_year = (df_homicidios_hechos['aaaa'] >= 2016) & (df_homicidios_hechos['aaaa'] <= 2021)
filter_data = df_homicidios_hechos[filter_year]

# Calculamos total de victimas
victims_data = filter_data.groupby('victima')['n_victimas'].sum().sort_values(ascending=False)

# Obtenemos el factor 'victima' con más victimas
victims_type = victims_data.idxmax()
print("Factor 'victima' con más Victimas:", victims_type)

# Obtenemos colores para el gráfico
x = len(victims_data)
color_selected = sns.color_palette("Blues", n_colors=x)

# Ordenamos las victimas por cantidad de victimas y diseñamos
sorted_victims = victims_data.sort_values(ascending=True)
color_select = {victims: color_selected[i] for i, victims in enumerate(sorted_victims.index)}

# Graficamos los datos
plt.figure(figsize=(16, 8))
bars = plt.bar(victims_data.index, victims_data, 
               color=[color_select[v] if v != victims_type else color_selected[-1] for v in victims_data.index])

# Caracteristicas del grafico
plt.title('Cantidad de Victimas x Tipo - Homicidios Hechos')
plt.xlabel('Tipo de Victima')
plt.ylabel('Numero de Victima')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Filtramos las filas que equivalga 'conductor'
driver=df_homicidios_victimas["rol"]=="conductor"

# Realizamos un agrupamiento y limpieza para poder obtener de una forma más clara la data
graph=df_homicidios_victimas[driver][["victima"]].groupby("victima").agg({"victima":"count"})
graph["cantidad"]=graph["victima"]
graph.drop(columns=["victima"],inplace=True)
graph.reset_index(inplace=True)

# Definimos colores
colores = plt.cm.Paired(range(len(graph)))

# Graficamos los datos
plt.figure(figsize=(6, 6))
plt.pie(graph['cantidad'], labels=graph['victima'], autopct='%1.1f%%', startangle=140, colors=colores)
plt.legend(loc='lower right')
plt.title('Visualización de Victimas por Cantidad - Homicidios Victimas')
plt.show()
```

```python
# TENER EN CUENTA QUE ESTE GRÁFICO SIRVE PARA CUALQUIER TIPO DE ACUSADO, SOLO ES CAMBIAR EL DATO PRINCIPAL
# Filtramos los datos relacionados a la columna 'victima'
bike_data = df_homicidios_hechos[df_homicidios_hechos['victima'] == 'moto']

# Contamos la frecuencia de acusados vinculados a 'moto' y ordenamos
acused_bikes = bike_data['acusado'].value_counts().sort_values(ascending=False).head(5)

# Creamos paleta de colores
n = len(acused_bikes)
selected_colors = sns.color_palette("Greens", n_colors=n)
color_select = {acusado: selected_colors[i] for i, acusado in enumerate(acused_bikes.index[::-1])}

# Creamos el grafico
plt.figure(figsize=(10, 6))
bars = plt.bar(acused_bikes.index, acused_bikes, 
               color=[color_select[acusado] for acusado in acused_bikes.index])

# Caracteristicas del grafico
plt.title('Acusados por Siniestro en Moto - Homicidios Hechos')
plt.xlabel('Acusado')
plt.ylabel('Cantidad de Incidentes')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Reemplazo los faltantes de las columnas pos x, pos y y XY(CABA) por 0 para poder utilizar estas columnas sin problemas para el analisis.
# Se reemplazan los valores faltantes por 0
df_homicidios_hechos['pos x'] = df_homicidios_hechos['pos x'].replace('.', 0)
df_homicidios_hechos['pos y'] = df_homicidios_hechos['pos y'].replace('.', 0)
df_homicidios_hechos['xy (caba)'] = df_homicidios_hechos['xy (caba)'].replace('Point (. .)', 0)
```

```python
# Modificamos para trabajar solo con la columna 'hh' y eliminamos la columna 'hora'.
df_homicidios_hechos.drop('hora', axis=1, inplace=True)
df_homicidios_hechos['hh'].apply(type).value_counts()
df_homicidios_hechos[df_homicidios_hechos['hh'].apply(type) == str]
```

```python
# Contamos la cantidad de victimas por comuna
df_homicidios_hechos['comuna'].sort_index()
recount_comuns = df_homicidios_hechos['comuna'].value_counts()

# Creamos el grafico
plt.figure(figsize=(8, 6))
plt.pie(recount_comuns, labels=recount_comuns.index, autopct='%1.1f%%', startangle=250, colors=['Green', 'Blue', 'Purple'])
plt.title('Distribucion x Comunas - Homicidios Hechos')
plt.axis('equal')
plt.show()
```

```python
# Realizamos un filtrado para con las edades de las victimas
rangos=range(1,110,10)
edad=df_homicidios_victimas["edad"]!="sd"
graph=df_homicidios_victimas[edad][["edad"]]
graph["edad"]=graph["edad"].apply(lambda x:int(x))
graph["rango"]=pd.cut(graph["edad"],bins=rangos,right=False)
graph=graph.groupby("rango").agg({'edad':"sum"})
graph.reset_index(inplace=True)

# Contamos la cantidad de homicidios
graph["rango"]=graph["rango"].apply(lambda x:str(x))

# Creamos el grafico
plt.figure(figsize=(15, 5))
plt.bar(graph['rango'], graph['edad'], color='green')

# Caracteristicas del grafico
plt.xticks(rotation=0)
plt.title('Edades de las Victimas - Homicidios Victimas')
plt.xlabel('Edad')
plt.ylabel('Cantidad')
plt.xticks(graph['rango'])

# Graficamos
plt.show()
```

```python
# Realizamos un filtrado para con los sexos de las victimas
victima_sexo = df_homicidios_victimas['sexo'].value_counts()

# Caracteristicas del grafico
plt.figure(figsize=(8, 6)) 
victima_sexo.plot(kind='bar', color='blue')
plt.title('Cantidad de Victimas por Sexo - Homicidios Victimas')
plt.xlabel('Sexo')  
plt.ylabel('Registros') 
plt.xticks(rotation=0) 
max_count = victima_sexo.max()
ylim_max = (int(max_count / 50) + 2) * 50
plt.ylim(0, ylim_max)

# Graficamos
plt.show()
```

```python
# TENER EN CUENTA QUE ESTE GRÁFICO SIRVE PARA CUALQUIER TIPO DE SEXO, SOLO ES CAMBIAR EL DATO PRINCIPAL
# Filtramos los registros de la columna 'sexo'
genero_victima = df_homicidios_victimas[df_homicidios_victimas['sexo'] == 'masculino']

# Contar la cantidad de registros por tipo de vehiculo
conteo_victima_vehiculo = genero_victima['victima'].value_counts()

# Caracteristicas del grafico
plt.figure(figsize=(10, 6))
conteo_victima_vehiculo.plot(kind='bar', color='purple') 
plt.title('Registros de Victimas Masculinas por Tipo de Vehiculo - Homicidios Victimas') 
plt.xlabel('Tipo de Vehiculo') 
plt.ylabel('Registros Masculinos')
plt.xticks(rotation=0)  

# Graficamos
plt.show()
```

```python
# Filtramos los registros por 'sexo' y cantidad de victimas por tipo de vehiculo
total_tipoVictima_sexo = df_homicidios_victimas.groupby(['sexo', 'victima']).size().unstack(fill_value=0)

# Caracteristicas del grafico
plt.figure(figsize=(10, 6)) 
total_tipoVictima_sexo.plot(kind='bar', stacked=True) 
plt.title('Cantidad de Tipo de Victima por Sexo - Homicidios Victimas') 
plt.xlabel('Tipo de Accidente') 
plt.ylabel('Cantidad de Accidentes') 
plt.xticks(rotation=45)  
plt.legend(title='Sexo', loc='upper right')  

# Graficamos
plt.show() 
```

```python
# Filtramos las filas que equivalga 'victima' a: 'auto' o 'moto'
victims_car_bike = df_lesiones[df_lesiones['victima'].isin(['moto', 'auto'])]

# Seleccionamos los años que deseamos marcar
selected_years = [2016, 2017, 2018, 2019, 2020, 2021]

# Filtramos filas por los años elegidos
selected_victims = victims_car_bike[victims_car_bike['aaaa'].isin(selected_years)]

# Calculamos la cantidad de victimas por tipo
victims_type = selected_victims.groupby('victima')['n_victimas'].sum()

# Graficamos los datos
plt.figure(figsize=(8, 8))
pie = plt.pie(victims_type, labels=victims_type.index, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Porcentaje de N° de Víctimas x Tipo - Lesiones')

# Encontramos indice del tipo de victima con mayor cantidad
index_max_victims = victims_type.idxmax()

# Asignamos colores
colores = ['purple' if tipo != index_max_victims else 'blue' for tipo in victims_type.index]

# Modificamos colores
for i, patch in enumerate(pie[0]):
    patch.set_color(colores[i])

plt.show()
```
