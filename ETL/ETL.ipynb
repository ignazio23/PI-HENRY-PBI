**Importación y Carga de Librerias**

```python
import pandas as pd
import numpy as np
```

**Carga de Datos y Análisis**

→ Para datos de 'homicidios'

```python
# Carga de los datos del archivo 'homicidios' del apartado 'hechos' y visualización de los mismos
df_homicidios_hechos = pd.read_excel('homicidios.xlsx',sheet_name = 'HECHOS')
df_homicidios_hechos
```

```python
# Visualización del tipo de datos
df_homicidios_hechos.info()
```

```python
# Convertimos los datos a minusculas
df_homicidios_hechos = df_homicidios_hechos.applymap(lambda x: x.lower() if isinstance(x, str) else x)

# Convertimos los titulos de las columnas a minusculas
df_homicidios_hechos.columns = df_homicidios_hechos.columns.str.lower()
```

```python
# Identificamos duplicados en las columnas
duplicates_hechos = df_homicidios_hechos[df_homicidios_hechos.duplicated(subset=['id', 'n_victimas', 'fecha', 'aaaa', 'mm', 'dd', 'hora', 'hh',
       'lugar_del_hecho', 'tipo_de_calle', 'calle', 'altura', 'cruce',
       'dirección normalizada', 'comuna', 'xy (caba)', 'pos x', 'pos y',
       'participantes', 'victima', 'acusado'])]

# Mostramos los datos duplicados
print(duplicates_hechos)
```

```python
# Verificamos valores nulos
nules_hechos = df_homicidios_hechos.isnull().any()

# Mostramos los valores nulos
print(nules_hechos)
```

```python
# Eliminamos los nulos
df_homicidios_hechos.dropna(how='all', inplace=True)
```

```python
# Cambiamos el tipo de dato de la columna 'altura'
df_homicidios_hechos['altura'] = df_homicidios_hechos['altura'].astype('object')

# Visualizamos el cambio
print(df_homicidios_hechos['altura'].dtype)
```

```python
# Reemplazamos los datos nulos por 'sd'
df_homicidios_hechos = df_homicidios_hechos.fillna(value='sd')

# Visualizamos los datos
df_homicidios_hechos
```

```python
# Convertimos el DF - df_homicidios_hechos a un CSV, y lo exportamos
df_homicidios_hechos.to_csv('../EDA/homicidios_hechos.csv', index=False)
```
